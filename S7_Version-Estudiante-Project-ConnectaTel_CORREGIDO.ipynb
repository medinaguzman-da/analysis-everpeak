### 2.3 Revisión y estandarización de fechas

**🎯 Objetivo:**  
Asegurar que las columnas de fecha estén correctamente formateadas y detectar años fuera de rango que indiquen errores de captura.

**Instrucciones:**  
- Convierte las columnas de fecha a tipo fecha y asegurate de que el código sea a prueba de errores.  
- Revisa cuántas veces aparece cada año.

Toma en cuenta que tenemos datos registrados hasta el año 2024.

```python

import pandas as pd
import numpy as np

# ==========================================
# 1. CONVERSIÓN DE FECHAS (Prueba de errores)
# ==========================================
# Convertimos a datetime manejando errores ('coerce' transformará formatos inválidos en NaT)
users['reg_date'] = pd.to_datetime(users['reg_date'], errors='coerce')
users['churn_date'] = pd.to_datetime(users['churn_date'], errors='coerce')
usage['date'] = pd.to_datetime(usage['date'], errors='coerce')

# Extraer el mes para los análisis mensuales indexados
users['month'] = users['reg_date'].dt.month
usage['month'] = usage['date'].dt.month

print("--- Tipos de datos post-conversión ---")
print(users[['reg_date', 'churn_date']].dtypes)
print(usage[['date']].dtypes)

# ==========================================
# 2. REVISIÓN DE DISTRIBUCIÓN TEMPORAL (Frecuencia de años)
# ==========================================
print("\n--- Distribución de años en Registro de Usuarios ---")
print(users['reg_date'].dt.year.value_counts(dropna=False))

print("\n--- Distribución de años en Dataset de Uso (Usage) ---")
print(usage['date'].dt.year.value_counts(dropna=False))

# ==========================================
# 3. CONTROL DE CALIDAD Y ANOMALÍAS EXPLICITAS
# ==========================================
# Frecuencia de tipos de consumo en usage (Solicitado por el revisor)
print("\n--- Distribución de tipos en usage['type'] ---")
print(usage['type'].value_counts(dropna=False))

# Conteo de valores nulos clave
print("\n--- Valores nulos detectados ---")
print(f"Nulos en churn_date (Usuarios activos): {users['churn_date'].isna().sum()}")
print(f"Nulos absolutos en dataset usage: \n{usage.isna().sum()}")

# Detección de anomalías en edad y ciudad
anomalous_age = users[users['age'] == -999].shape[0]
anomalous_city = users[users['city'] == '?'].shape[0]

print(f"\nRegistros con edad inválida (-999): {anomalous_age}")
print(f"Registros con ciudad inválida ('?'): {anomalous_city}")
```

- Se detectó que las columnas de fechas originales (`reg_date` y `churn_date` en el dataset de usuarios) estaban cargadas inicialmente como tipo texto (`object`).
- Se aplicó la función `pd.to_datetime()` para transformarlas al formato correcto de fecha y hora (`datetime64`). El parámetro `errors='coerce'` garantiza que cualquier valor con formato inválido se convierta automáticamente en `NaT`, haciendo el proceso a prueba de errores.
- Se creó una nueva columna `month` (mes) extrayendo el componente mensual de `reg_date` en `users` y de `date` en `usage`, lo cual permite agrupar las métricas de consumo de manera mensual por cliente.

## 🧩Paso 3: Limpieza básica de datos

### 3.1 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Aplicar reglas de limpieza para reemplazar valores sentinels y corregir fechas imposibles.

**Instrucciones:**  
- En `age`, reemplaza el sentinel **-999** con la mediana.
- En `city`, reemplaza el sentinel `"?"` por valores nulos (`pd.NA`).  
- Marca como nulas (`pd.NA`) las fechas fuera de rango.

```python

import pandas as pd
import numpy as np

# =====================================================================
# 1. IMPUTACIÓN DE EDAD (Sentinel -999 reemplazado por la mediana)
# =====================================================================
age_mediana = users.loc[users['age'] != -999, 'age'].median()
users['age'] = users['age'].replace(-999, age_mediana)

print("--- Verificación de Edad ---")
print(users['age'].describe())


# =====================================================================
# 2. LIMPIEZA DE CIUDAD (Sentinel '?' reemplazado por nulo nativo pd.NA)
# =====================================================================
users['city'] = users['city'].replace('?', pd.NA)

print("\n--- Verificación de Ciudades Limpias ---")
print(f"Nulos en 'city': {users['city'].isna().sum()}")


# =====================================================================
# 3. CORRECCIÓN DE FECHAS IMPOSIBLES (Límite estricto: Año 2024)
# =====================================================================
# Definimos el límite operativo real del proyecto como un objeto Timestamp de Pandas
fecha_limite = pd.Timestamp('2024-12-31')

# Forzamos cualquier fecha posterior al 2024 a convertirse en NaT (Not a Time)
users.loc[users['reg_date'] > fecha_limite, 'reg_date'] = pd.NaT

print("\n--- Verificación de Fechas Fuera de Rango ---")
print(f"Fechas futuras detectadas y convertidas a NaT: {users['reg_date'].isna().sum()}")
```

### 3.2 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Decidir qué hacer con los valores nulos según su proporción y relevancia.

**Instrucciones:**
- Verifica si los nulos en `duration` y `length` son **MAR**(Missing At Random) revisando si dependen de la columna `type`.  
  Si confirmas que son MAR, **déjalos como nulos** y justifica la decisión.

```python
# Verificación MAR en usage (Missing At Random) para duration
pd.crosstab(usage['type'], usage['duration'].isna(), normalize='index')
```

```python
# Verificación MAR en usage (Missing At Random) para length
pd.crosstab(usage['type'], usage['length'].isna(), normalize='index')
```

COMENTARIOS:

- Consistencia en duration: La variable de duración es completamente válida para las llamadas (call, 0% de nulos), pero es nula en el 99.9% de los mensajes de texto (text). Esto es bioconsistente con el negocio, ya que los SMS no poseen una duración medible en minutos.

🎯 **Objetivo:**  
Detectar valores extremos en las variables clave de **uso** y **clientes** que podrían afectar el análisis, y decidir si requieren limpieza o revisión adicional.

**Instrucciones:**

  - `age` 
  - `cant_mensajes`
  - `cant_llamadas`
  - `total_minutos_llamada`  
- Crea un **for** para generar los 4 boxplots automáticamente.
<br>

- Después de crear los gráfico, responde si **existen o no outliers** en las variables.  
- Si hay outliers, crea otro bucle para calcular los límites de esas columnas usando el **método IQR** y decide qué hacer con ellos.
  - Si solamente hay outliers de un solo lado, no es necesario calcular ambos límites.

**Hint:**
- Dentro del bucle, usa `plt.title(f'Boxplot: {col}')` para que el título cambie acorde a la columna.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir la lista de columnas numéricas de interés
columnas_interes = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

# Configurar el tamaño del lienzo para los gráficos
plt.figure(figsize=(15, 10))

# Bucle for para generar los 4 boxplots automáticamente
for i, col in enumerate(columnas_interes, 1):
    plt.subplot(2, 2, i)  # Crea una cuadrícula de 2x2
    sns.boxplot(x=user_profile[col], color='skyblue')
    plt.title(f'Boxplot: {col}')  # Título dinámico solicitado en el Hint
    plt.xlabel(col)
    plt.grid(axis='x', linestyle='--', alpha=0.7)

# Ajustar los espacios entre subplots y mostrar el lienzo completo
plt.tight_layout()
plt.show()
```

💡Insights: 
- Hay una relación fuerte y positiva entre variables de consumo: Existe una correlación extremadamente alta y positiva entre cant_llamadas y cant_minutos_llamada (r = 0.96). Esto indica una relación lineal casi perfecta, a mayor cantidad de llamadas realizadas por un usuario, mayor es el acumulado total de minutos consumidos. Ambas métricas se mueven de la mano de forma lógica.
- Hay una relación moderada-fuerte entre llamadas y mensajes: La correlación entre cant_mensajes y las variables de llamadas (cant_llamadas con r = 0.63 y cant_minutos_llamada con r = 0.61) es moderadamente fuerte y positiva. Esto demuestra que los usuarios que tienden a ser muy activos enviando mensajes de texto también suelen tener un consumo elevado en los servicios de voz/llamadas. No son canales excluyentes, sino complementarios para el usuario intensivo.
- Hay una ausencia de correlación con la edad: La variable age (edad) presenta valores prácticamente cercanos a cero con todas las demás métricas de consumo (cant_mensajes: -0.02, cant_llamadas: -0.04, cant_minutos_llamada: -0.04). Esto revela de forma contundente que la edad no influye en absoluto en el comportamiento de uso del servicio, por lo que los jóvenes y adultos mayores consumen minutos y mensajes de forma indistinta.

```python

# Definir la lista de columnas de interés
columnas_limites = ['age', 'cant_mensajes', 'cant_llamadas', 'cant_minutos_llamada']

# Bucle para calcular y mostrar los límites IQR por columna
for col in columnas_limites:
    # Calcular los cuartiles Q1 y Q3
    q1 = user_profile[col].quantile(0.25)
    q3 = user_profile[col].quantile(0.75)
    
    # Calcular el Rango Intercuartílico (IQR)
    iqr = q3 - q1
    
    # Calcular los límites teórico-estadísticos
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    # Imprimir los resultados de forma clara
    print(f"--- Límites para la columna: {col} ---")
    print(f"  Q1 (25%): {q1:.2f} | Q3 (75%): {q3:.2f} | IQR: {iqr:.2f}")
    print(f"  Límite Inferior: {limite_inferior:.2f}")
    print(f"  Límite Superior: {limite_superior:.2f}\n")
```

```python

# Revisa los limites superiores y el max, para tomar la decisión de mantener los outliers o no
user_profile[columnas_limites].describe()
```

### Análisis ejecutivo

⚠️ **Problemas detectados en los datos**
- **Valores nulos y caracteres de ruido en `city`:** La columna de ciudades presentó un 11.73% de valores faltantes (469 registros) más 96 registros con el carácter erróneo `"?"`. Ambos fueron consolidados como nulos nativos (`pd.NA`), sin asignar etiquetas de reemplazo como "Desconocida", con el fin de preservar la pureza de las distribuciones geográficas y reflejar fielmente la ausencia de información en el sistema de origen.
- **Valores sentinel en `age`:** Se detectó el valor `-999` en la columna de edad, un código de error de captura que distorsionaba la media y la desviación estándar reales. Se corrigió imputando estos registros con la mediana muestral calculada sobre los valores válidos.
- **Inconsistencias de formato en fechas:** Las columnas `reg_date` y `churn_date` estaban cargadas como texto plano (`object`) y se convirtieron a tipo `datetime64`. Se estableció el límite operativo estricto de `2024-12-31`, marcando como `NaT` cualquier fecha que superara ese umbral.
- **Campos vacíos interpretables en `churn_date`:** El 88.35% de registros (3,534 filas) presentó valores ausentes. La auditoría confirmó que esto no es un error, sino que representa a los usuarios activos que no han dado de baja su servicio; por tanto, se mantuvieron como nulos.


🔍 **Segmentos por Edad**
- **Segmento Joven (< 30 años):** Grupo minoritario del negocio con aproximadamente 750 usuarios. Muestra un comportamiento de consumo similar al resto de los segmentos, ya que la edad no correlaciona con el nivel de uso.
- **Segmento Adulto (30 a 59 años):** Pilar comercial dominante, concentrando a más de 2,000 usuarios. Al ser el grupo más numeroso y distribuirse entre todos los niveles de uso, representa la mayor oportunidad de upselling hacia el plan Premium.
- **Segmento Adulto Mayor (≥ 60 años):** Alrededor de 1,250 usuarios con patrones de consumo similares a los demás grupos, lo que refuerza que la oferta de planes no debe segmentarse por edad.


📊 **Segmentos por Nivel de Uso**
- **Uso Medio:** Núcleo central con casi 3,000 usuarios. Consumo regular y moderado; son los mejores candidatos para un plan intermedio que llene la brecha entre Básico y Premium.
- **Bajo Uso:** Aproximadamente 800 usuarios con consumos mínimos (menos de 5 llamadas y mensajes). Requieren estrategias de retención o planes de bajo costo para evitar churn.
- **Alto Uso:** Nicho de alto valor con unos 250 clientes de consumo intensivo. Son candidatos directos para migración al plan Premium y para ofertas corporativas dedicadas.


El análisis de correlación demostró que la edad no influye en el consumo ($r \approx 0.0$), por lo que los segmentos de uso son el criterio definitorio para la estrategia comercial. La concentración masiva en el plan Básico (~65%) y en el nivel de Uso Medio evidencia una brecha comercial significativa: muchos usuarios con consumo creciente permanecen en el plan de menor costo porque no existe una opción intermedia que los incentive a migrar.


💡 **Recomendaciones**
- **Upselling dirigido al segmento de Alto Uso:** Identificar automáticamente a los clientes del plan Básico clasificados como "Alto Uso" (aproximadamente 250 usuarios con más de 10 llamadas y mensajes mensuales) y ofrecerles activamente la migración al plan Premium, mostrando que su consumo real supera los límites económicos de su suscripción actual.
- **Creación de un plan intermedio para el segmento de Uso Medio:** Diseñar un "Plan Conecta" o "Premium Lite" orientado a los casi 3,000 usuarios del segmento de Uso Medio, cerrando la brecha comercial entre el plan Básico y el Premium actual. Este segmento es el más poblado y el que mayor potencial de monetización incremental tiene.
- **Retención preventiva para el segmento de Bajo Uso:** Implementar micro-paquetes o tarifas de mantenimiento de bajo costo para los ~800 usuarios del segmento de Bajo Uso, reduciendo el riesgo de churn hacia competidores con ofertas más agresivas y asegurando ingresos recurrentes mínimos sin saturar la red.
